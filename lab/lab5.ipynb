```python
def clean_text(text):   
    clean_text = re.sub(r'[^A-Za-zÀ-ỹ0-9\s]', '', text)
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    return clean_text
```

```python
words_list = text.split('\n')
```

```python
words_list
```

```python
for i in range(len(words_list)):
    words_list[i] = clean_text(words_list[i])

for i in range(len(words_list)):
    if words_list[i] == '':
        words_list.remove(words_list[i])
```

```python
with open("words_list.json", "w", encoding="utf-8") as f:
    f.write("{\n")
    
    f.write("}")
```

```python
dict = {"id": [], "text": []}
for index, word in enumerate(words_list):
    dict["id"].append(index+1)
    dict["text"].append(word)
```

```python
sr = pd.Series(dict["text"], index=dict["id"])
sr
```

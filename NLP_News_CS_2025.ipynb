```python
import pandas as pd
from underthesea import word_tokenize

from collections import Counter
import re

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.preprocessing import OrdinalEncoder
```

```python
def pipeline(X, y, test_size: float = 0.2):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    # Logistic Regression
    log_model = LogisticRegression(max_iter=1000)
    log_model.fit(X_train, y_train)
    pred_log = log_model.predict(X_test)

    # SVM
    svm_model = LinearSVC()
    svm_model.fit(X_train, y_train)
    pred_svm = svm_model.predict(X_test)

    # Naive Bayes
    nb_model = MultinomialNB()
    nb_model.fit(X_train, y_train)
    pred_nb = nb_model.predict(X_test)

    print("Logistic Regression:\n", classification_report(y_test, pred_log))
    print("-" * 80, end="\n\n")
    print("SVM:\n", classification_report(y_test, pred_svm))
    print("-" * 80, end="\n\n")
    print("Naive Bayes:\n", classification_report(y_test, pred_nb))
```

# 1. 

```python
# Gộp title và content
combined_df["text"] = combined_df["title"] + "|" + combined_df["content"]

# Tách từ (tokenize bằng dấu '|') và làm sạch
combined_df["tokens"] = combined_df["text"].apply(lambda x: [t.strip() for t in x.split("|") if t.strip()])

# Gộp theo topic: gom tất cả token của các id cùng topic
topic_tokens = combined_df.groupby("topic")["tokens"].sum()

# Đếm tần suất từ mỗi topic
topic_word_counts = {topic: Counter(tokens) for topic, tokens in topic_tokens.items()}

# Biến thành DataFrame (Bag of Words matrix)
data = pd.DataFrame(topic_word_counts).fillna(0).astype(int).T  # mỗi hàng = topic, mỗi cột = từ

data = data.reset_index().rename(columns={"index": "topic"})

data
```

```python
# Tạo DataFrame đặc trưng
X = data.drop(columns=['topic'])
y = data['topic']

print(X.shape)
print(y.shape)
```

```python
y
```

```python
pipeline(X, y)
```

# 2

```python
# Gộp và tách từ
combined_df["text"] = combined_df["title"] + "|" + combined_df["content"]
combined_df["tokens"] = combined_df["text"].apply(lambda x: [t.strip() for t in x.split("|") if t.strip()])

# Gom token theo topic
topic_tokens = combined_df.groupby("topic")["tokens"].sum()

# Tạo set từ cho mỗi topic
topic_word_sets = {topic: set(tokens) for topic, tokens in topic_tokens.items()}

# Tìm giao các từ giữa các topic
common_words = set()
topics = list(topic_word_sets.keys())
for i in range(len(topics)):
    for j in range(i+1, len(topics)):
        common_words |= (topic_word_sets[topics[i]] & topic_word_sets[topics[j]])

# Loại bỏ từ chung ra khỏi từng topic
unique_topic_tokens = {topic: [w for w in tokens if w not in common_words]
                       for topic, tokens in topic_tokens.items()}

# Đếm tần suất trong từ còn lại
unique_word_counts = {topic: Counter(words) for topic, words in unique_topic_tokens.items()}

# Biến thành DataFrame
filtered_bow_df = pd.DataFrame(unique_word_counts).fillna(0).astype(int).T

filtered_bow_df = filtered_bow_df.reset_index().rename(columns={"index": "topic"})

filtered_bow_df
```

```python
X2 = filtered_bow_df.drop(columns=['topic'])
y2 = filtered_bow_df['topic']
```

```python
X2.iloc[:, 2000:3000]
```

```python
pipeline(X2, y2)
```

```python
import pandas as pd
from underthesea import word_tokenize

from collections import Counter
import re

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame()
```

```python
for i in range (1,12):
    df_temp = pd.read_excel(r".\assets\nlp_class_dataset\G01.xlsx".replace("01", str(i).zfill(2)))
    df = pd.concat([df, df_temp], ignore_index=True)
```

# Demand

1. Tạo một bảng dữ liệu trong đó, mỗi từ là một feature, và giá trị của nó là số lần xuất hiện của nó. Có thể cộng cả content+title lại. Tiếp đến feed bảng dữ liệu này vào 3 mô hình học máy của scikit (Logistic, SVM, NaiveBayes)

2. Làm tương tự ở 1, nhưng bảng dữ liệu với số thuộc tính giảm sau khi loại các từ xuất hiện chung trong nhiều nhóm (có thể sử dụng set kết hợp với intersect - phép giao). Bảng features này có kích thước nhỏ hơn nhiều so với bảng 1.

3. Không tạo bảng features như 1 và 2 mà sử dụng kĩ thuật BoW (trong scikit, sử dụng CountVectorize)

4. Không tạo bảng feature như 1 và 2 mà sử dụng tfidf (TfidfVectorizer của scikit)

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.preprocessing import OrdinalEncoder
```

```python
def pipeline(X, y, test_size: float = 0.2):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    # Logistic Regression
    log_model = LogisticRegression(max_iter=1000)
    log_model.fit(X_train, y_train)
    pred_log = log_model.predict(X_test)

    # SVM
    svm_model = LinearSVC()
    svm_model.fit(X_train, y_train)
    pred_svm = svm_model.predict(X_test)

    # Naive Bayes
    nb_model = MultinomialNB()
    nb_model.fit(X_train, y_train)
    pred_nb = nb_model.predict(X_test)

    print("Logistic Regression:\n", classification_report(y_test, pred_log))
    print("-" * 80, end="\n\n")
    print("SVM:\n", classification_report(y_test, pred_svm))
    print("-" * 80, end="\n\n")
    print("Naive Bayes:\n", classification_report(y_test, pred_nb))
```

```python
words_list = []
for i in range(len(combined_df)):
    words_list.extend(combined_df.loc[i, "title"].split(" | "))
    words_list.extend(combined_df.loc[i, "content"].split(" | "))
    
word_freq = pd.Series(words_list).value_counts()
word_freq
```

```python
top_k = len(word_freq.index) 
vocab = word_freq.head(top_k).index.tolist()

rows = []
for i in range(len(combined_df)):
    tokens = []
    tokens.extend([w for w in combined_df.loc[i, "title"].split(" | ") if w])
    tokens.extend([w for w in combined_df.loc[i, "content"].split(" | ") if w])
    cnt = Counter(tokens)
    rows.append({w: cnt.get(w, 0) for w in vocab})

data = pd.DataFrame(rows)
data.insert(0, "topic", combined_df["topic"].values)

data.shape
```

```python
data
```

```python
X = data.drop(columns=["topic"])
y = data["topic"]
```

```python
pipeline(X, y)
```

```python
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())

# Tạo cột text nếu chưa có
df['text'] = df['title'] + ' ' + df['content']

# Tạo danh sách từ theo nhóm label
group_words = {}
for label, group in df.groupby('topic'):
    words = set()
    for text in group['text']:
        words.update(tokenize(text))
    group_words[label] = words

# Tìm các từ xuất hiện ở nhiều hơn 1 nhóm
common_words = set()
labels = list(group_words.keys())
for i in range(len(labels)):
    for j in range(i+1, len(labels)):
        common_words.update(group_words[labels[i]].intersection(group_words[labels[j]]))

# Tạo tập từ riêng (exclusive) cho mỗi nhóm
exclusive_words = set()
for label, words in group_words.items():
    exclusive_words.update(words - common_words)

print(f"Tổng số từ ban đầu: {len(set.union(*group_words.values()))}")
print(f"Tổng số từ sau khi loại từ chung: {len(exclusive_words)}")
```

```python
# Hàm tokenize
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())

# Gộp text
df['text'] = df['title'] + ' ' + df['content']

# Gom nhóm theo topic
group_words = {}
for label, group in df.groupby('topic'):
    words = set()
    for text in group['text']:
        words.update(tokenize(text))
    group_words[label] = words

# Tìm các từ xuất hiện ở nhiều hơn 1 nhóm
common_words = set()
labels = list(group_words.keys())
for i in range(len(labels)):
    for j in range(i + 1, len(labels)):
        common_words.update(group_words[labels[i]].intersection(group_words[labels[j]]))

# Giữ lại các từ độc quyền
exclusive_words = set()
for label, words in group_words.items():
    exclusive_words.update(words - common_words)
```

```python
# Tạo feature
features_reduced = []
for t in df['text']:
    counts = Counter(tokenize(t))
    row = {word: counts.get(word, 0) for word in exclusive_words}
    features_reduced.append(row)

X2 = pd.DataFrame(features_reduced)
y2 = df['topic']

X2 = X2.reset_index(drop=True)
y2 = y2.reset_index(drop=True)

print(X2.shape)
print(y2.shape)
```

```python
X2.head()
```

```python
pipeline(X2, y2)
```

```python
# Gộp title + content
df['text'] = df['title'] + ' ' + df['content']

# Khởi tạo CountVectorizer (Bag of Words)
vectorizer = CountVectorizer()

# Fit và transform dữ liệu văn bản -> ma trận đặc trưng
X3 = vectorizer.fit_transform(df['text'])

# Nhãn
y3 = df['topic']

print(len(vectorizer.get_feature_names_out()))
print(X3.shape)
```

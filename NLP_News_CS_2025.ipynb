```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.preprocessing import OrdinalEncoder
```

```python
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())
```

```python
def pipeline(X, y, test_size: float = 0.2):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    # Logistic Regression
    log_model = LogisticRegression(max_iter=1000)
    log_model.fit(X_train, y_train)
    pred_log = log_model.predict(X_test)

    # SVM
    svm_model = LinearSVC()
    svm_model.fit(X_train, y_train)
    pred_svm = svm_model.predict(X_test)

    # Naive Bayes
    nb_model = MultinomialNB()
    nb_model.fit(X_train, y_train)
    pred_nb = nb_model.predict(X_test)

    print("Logistic Regression:\n", classification_report(y_test, pred_log))
    print("-" * 80, end="\n\n")
    print("SVM:\n", classification_report(y_test, pred_svm))
    print("-" * 80, end="\n\n")
    print("Naive Bayes:\n", classification_report(y_test, pred_nb))
```

# 1. 

```python
# Gộp nội dung
df['text'] = df['title'].fillna('') + ' ' + df['content'].fillna('')

# Hàm tách từ cơ bản
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())

# Tạo tập hợp tất cả các từ trong toàn bộ corpus
all_words = set()
for t in df['text']:
    all_words.update(tokenize(t))

print("Tổng số từ duy nhất trong tập:", len(all_words))
```

```python
# Tạo danh sách từ điển cho từng bài
features = []
for text in df['text']:
    counts = Counter(tokenize(text))
    row = {word: counts.get(word, 0) for word in all_words}
    features.append(row)

# Tạo DataFrame đặc trưng
X = pd.DataFrame(features).fillna(0)
y = df['topic']

print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
pipeline(X, y)
```

```python
# Tạo cột text nếu chưa có
df['text'] = df['title'] + ' ' + df['content']

# Tạo danh sách từ theo nhóm label
group_words = {}
for label, group in df.groupby('topic'):
    words = set()
    for text in group['text']:
        words.update(tokenize(text))
    group_words[label] = words

# Tìm các từ xuất hiện ở nhiều hơn 1 nhóm
common_words = set()
labels = list(group_words.keys())
for i in range(len(labels)):
    for j in range(i+1, len(labels)):
        common_words.update(group_words[labels[i]].intersection(group_words[labels[j]]))

# Tạo tập từ riêng (exclusive) cho mỗi nhóm
exclusive_words = set()
for label, words in group_words.items():
    exclusive_words.update(words - common_words)

print(f"Tổng số từ ban đầu: {len(set.union(*group_words.values()))}")
print(f"Tổng số từ sau khi loại từ chung: {len(exclusive_words)}")
```

```python
def normalize_tokens(text: str):
    # tách từ tiếng Việt rồi chuẩn hóa
    toks = word_tokenize(text)
    cleaned = []
    for t in toks:
        t = t.lower().strip()
        # bỏ ký tự không phải chữ/số (giữ chữ cái tiếng Việt)
        t = re.sub(r'[^a-zà-ỹ0-9_]+', '', t)
        if t:
            cleaned.append(t)
    return cleaned
```

```python
# Tạo tập từ theo topic từ df gốc
group_words = {}
for topic, group in df.groupby('topic'):
    topic_vocab = set()
    for doc in group['text']:
        topic_vocab.update(normalize_tokens(doc))
    group_words[topic] = topic_vocab

# Xác định các từ xuất hiện ở >= 2 topic
word_topic_count = {}
for topic, vocab in group_words.items():
    for w in vocab:
        word_topic_count[w] = word_topic_count.get(w, 0) + 1

common_words = {w for w, c in word_topic_count.items() if c >= 2}

# Tạo từ độc quyền theo từng topic
exclusive_by_topic = {
    topic: vocab - common_words
    for topic, vocab in group_words.items()
}

# Hợp nhất tất cả từ độc quyền
exclusive_words = set().union(*exclusive_by_topic.values())

print(f"Tổng số từ khác nhau (toàn bộ): {len(set().union(*group_words.values()))}")
print(f"Tổng số từ độc quyền sau khi loại: {len(exclusive_words)}")
```

```python
# Tạo feature
features_reduced = []
for t in df['text']:
    counts = Counter(tokenize(t))
    row = {word: counts.get(word, 0) for word in exclusive_words}
    features_reduced.append(row)

X2 = pd.DataFrame(features_reduced)
y2 = df['topic']

X2 = X2.reset_index(drop=True)
y2 = y2.reset_index(drop=True)

print(X2.shape)
print(y2.shape)
```

```python
X2.head()
```

```python
pipeline(X2, y2)
```
